### Analysis

- Capomulin & Ramicane had the highest number of mice tested versus other treatements.  These relatively larger sample sizes indicate higher trust in output metrics for these particular treatments.
- The ratio of male to female mice was nearly 1:1.  Any skewed data should not be a result of the mice sex since each group is represented near equally.  
- Capomulin & Ramicane are clearly, and closely, the top promising treatments.  Relative to the other two most promising treatments (Infubinol & Ceftamin), these treatments produced significant reduced tumor size at the final Time Point.  This is statistically validated as the box plots show large deviation between means & quartiles that is not driven by outliers. 
- Viewing single mice on Capomulin shows variance in when certain mice respond.  I.e. some mice responded in a more consistent linear tumor size reduction, whereas others had more volatile tumor size over time, while some mice tumor size did not respond until later Time Points.  Ultimately, mice treated with Capomulin all showed a significant reduction in tumor size at end of measured treatement period.
- Linear regression of Mouse Weight vs Avg Tumor Size showed a consistent linear correlation with an r-value of 0.84.  This r-value is high enough to place a higher level of trust in the correlation.  As a result of this particular analysis, we can consider utilizing Mouse Weight as a simple proxy for tumor size during treatment. 

 

```python
# Dependencies and Setup
# %matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem

# Study data files
mouse_metadata_path = "./data/Mouse_metadata.csv"
study_results_path = "./data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single DataFrame
combined_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
combined_df
```

```python
# Checking the number of mice.
len(combined_df["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
# dup_mouse = combined_df[combined_df.duplicated(subset=["Mouse ID", "Timepoint"])]
dup_mouse = combined_df[combined_df.duplicated(subset=["Mouse ID", "Timepoint"])]["Mouse ID"].unique()
dup_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID.
combined_df[ (combined_df["Mouse ID"] == "g989" )  ]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# combined_df[ combined_df["Mouse ID"].isin(dup_mouse) == False ]
clean_df = combined_df[ combined_df["Mouse ID"] != "g989" ]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
```

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
means = clean_df.groupby("Drug Regimen").mean(numeric_only=True)["Tumor Volume (mm3)"]
# means
medians = clean_df.groupby("Drug Regimen").median(numeric_only=True)["Tumor Volume (mm3)"]
# medians
variances = clean_df.groupby("Drug Regimen").var(numeric_only=True)["Tumor Volume (mm3)"]
# variances
stddevs = clean_df.groupby("Drug Regimen").std(numeric_only=True)["Tumor Volume (mm3)"]
# stddevs
SEMs = clean_df.groupby("Drug Regimen").sem(numeric_only=True)["Tumor Volume (mm3)"]
# SEMs
```

```python
# Assemble the resulting series into a single summary DataFrame.
pd.DataFrame({"Mean Tumor Volume": means,
              "Median Tumor Volume": medians,
              "Variance Tumor Volume": variances,
              "Standard Deviation Tumor Volume": stddevs,
              "Standard Error Tumor Volume": SEMs })
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
clean_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean", "median", "var", "std", "sem"]})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
```

```python
# Start by getting the last (greatest) timepoint for each mouse
max_tumor = clean_df.groupby("Mouse ID")["Timepoint"].max()
max_tumor = max_tumor.reset_index()
max_tumor
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
# final_timepoint_df = max_tumor.merge(clean_df, on=["Mouse ID", "Timepoint"], how="left")
final_timepoint_df = pd.merge(max_tumor, clean_df, on=["Mouse ID", "Timepoint"], how="left")
final_timepoint_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Capomulin, Ramicane, Infubinol, and Ceftamin

# clean_df["Drug Regimen"].unique()

# treatments
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []
# tumor_vol_data

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for drug in treatments:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = final_timepoint_df.loc[final_timepoint_df["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    # print(final_tumor_vol)

    # add subset
    tumor_vol_data.append(final_tumor_vol)
    # print(tumor_vol_data)

    # Determine outliers using upper and lower bounds
    quartiles = final_tumor_vol.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)

    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    # print(outliers)
    print(f"{drug}'s potential outliers {outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
outlier_format = dict(markerfacecolor='red', markersize=10)
plt.boxplot(tumor_vol_data, labels=treatments , flierprops=outlier_format)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"])
# correlation
print(f"The correlation between Mouse Weight and Avg Tumor Volume is {round(correlation[0],2)}")
```

```python
# Calculate linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
linreg = st.linregress(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"])
linreg
```

```python
slope = linreg[0]
# slope
intercept = linreg[1]
# intercept
linereg_y_values = capomulin_avg["Weight (g)"] * slope + intercept
```

```python
plt.scatter(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"])
plt.plot(capomulin_avg["Weight (g)"], linereg_y_values, color="red")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Avg Tumor Volume (mm3)")
plt.title(f"Capomulin treatment: Linear Regression, Weight vs Avg Tumor Vol")
plt.show()
```
